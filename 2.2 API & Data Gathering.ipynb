#### 5. Create weather df

```python
# Convert to json file 

date_str = json.loads(r.text) 
```

```python
date_str
```

```python
# Secure all items in the response that correspond to TAVG

avg_temps = [item for item in date_str ['results'] if item['datatype']=='TAVG']
```

```python
# Get only the date field from all average temperature readings

dates_temp = [item['date'] for item in avg_temps]
```

```python
# Get the temperature from all average temperature readings

temps = [item['value'] for item in avg_temps]
```

```python
temps
```

```python
# Put results in df

df_temp = pd.DataFrame()
```

```python
# Convert dates so only the date is visible 

df_temp['date'] = [datetime.strptime(date_str, "%Y-%m-%dT%H:%M:%S") for date_str in dates_temp]
df_temp['avgTemp'] = [float(v)/10.0 for v in temps] 
```

#### 6. Merge Dataframes

```python
# Convert "started_at" to date format 

df_combined['started_at'] = pd.to_datetime(df_combined['started_at'], dayfirst = True)
df_combined['date'] = pd.to_datetime(df_combined['started_at'], format='%Y-%m-%d').dt.date 
```

```python
# Convert "date" to datetime

df_combined['date'] = pd.to_datetime(df_combined['date'])
df_temp['date'] = pd.to_datetime(df_temp['date'])
```
